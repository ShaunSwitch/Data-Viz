```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
import seaborn as sns
from matplotlib import rcParams
eonClr = ["#FF0000","#E3E000","#5CC1CB","#b00402","#A66999","#3599B8","#DFBFBF","#4AC5BB","#5F6B6D","#FB8281","#F4D25A","#7F898A","#A4DDEE","#FDAB89","#B687AC","#28738A","#A78F8F","#168980","#293537","#BB4A4A","#B59525","#475052","#6A9FB0","#BD7150","#7B4F71","#1B4D5C","#706060","#0F5C55","#1C2325"]
sns.set_palette(sns.color_palette(eonClr))
rcParams['figure.figsize'] = 16,10
job_hist = pd.read_csv(r"data_viz.csv",encoding = "ISO-8859-1")
job_hist.describe()
```

```python
job_hist = job_hist[~(job_hist['WORK_DURATION'] < 0)]
job_hist = job_hist[~(job_hist['ACTUAL_TRAVEL_TIME'] < 0)]
job_hist = job_hist[job_hist['WORK_TYPE']=='SMART']
```

```python
job_hist['COMPLETE_COUNT'] = 1
job_hist['ABORT_COUNT'] = 1
job_hist['COMPLETE_COUNT'] = job_hist.apply(lambda x: x['COMPLETE_COUNT'] if x['WORK_STATUS4']=='COMPLETED' else np.nan, axis=1)
job_hist['ABORT_COUNT'] = job_hist.apply(lambda x: x['ABORT_COUNT'] if x['WORK_STATUS4'] in ['CUSTOMER REFUSAL ABORT','TECHNICAL ABORT','NO ACCESS ABORT','INSTALL_AND_LEAVE','REVERT_TO_CLASSIC'] else np.nan, axis=1)
job_hist['APPOINTMENT_DATE'] = pd.to_datetime(job_hist['APPOINTMENT_DATE'], format='%d/%m/%Y')
job_hist['JOB_WEEK_DATE'] = job_hist.APPOINTMENT_DATE.dt.to_period('W').apply(lambda r: r.start_time)
job_hist['JOB_MONTH_DATE'] = pd.to_datetime(job_hist.APPOINTMENT_DATE).dt.to_period('M').dt.to_timestamp()
job_hist.head()

```

```python
mnth_df_time = pd.pivot_table(data = job_hist[(job_hist['WORK_STATUS4'] == 'COMPLETED')&(job_hist['FUEL_TYPE'] == 'DUAL FUEL')], values = ['WORK_DURATION'],index = ['JOB_MONTH_DATE'],aggfunc=np.mean)
mnth_elec_time = pd.pivot_table(data = job_hist[(job_hist['WORK_STATUS4'] == 'COMPLETED')&(job_hist['FUEL_TYPE'] == 'ELECTRICITY')], values = ['WORK_DURATION'],index = ['JOB_MONTH_DATE'],aggfunc=np.mean)
mnth_gas_time = pd.pivot_table(data = job_hist[(job_hist['WORK_STATUS4'] == 'COMPLETED')&(job_hist['FUEL_TYPE'] == 'GAS')], values = ['WORK_DURATION'],index = ['JOB_MONTH_DATE'],aggfunc=np.mean)
mnth_travel_time = pd.pivot_table(data = job_hist[(job_hist['WORK_STATUS4'] == 'COMPLETED')], values = ['ACTUAL_TRAVEL_TIME'],index = ['JOB_MONTH_DATE'],aggfunc=np.mean)
mnth_comp_df = pd.pivot_table(data = job_hist[(job_hist['WORK_STATUS4'].isin(['CUSTOMER REFUSAL ABORT','TECHNICAL ABORT','NO ACCESS ABORT','INSTALL_AND_LEAVE','REVERT_TO_CLASSIC','COMPLETED']))&(job_hist['FUEL_TYPE'] == 'DUAL FUEL')], values = ['COMPLETE_COUNT','ABORT_COUNT'],index = ['JOB_MONTH_DATE'],aggfunc=np.sum)
mnth_comp_elec = pd.pivot_table(data = job_hist[(job_hist['WORK_STATUS4'].isin(['CUSTOMER REFUSAL ABORT','TECHNICAL ABORT','NO ACCESS ABORT','INSTALL_AND_LEAVE','REVERT_TO_CLASSIC','COMPLETED']))&(job_hist['FUEL_TYPE'] == 'ELECTRICITY')], values = ['COMPLETE_COUNT','ABORT_COUNT'],index = ['JOB_MONTH_DATE'],aggfunc=np.sum)
mnth_comp_gas = pd.pivot_table(data = job_hist[(job_hist['WORK_STATUS4'].isin(['CUSTOMER REFUSAL ABORT','TECHNICAL ABORT','NO ACCESS ABORT','INSTALL_AND_LEAVE','REVERT_TO_CLASSIC','COMPLETED']))&(job_hist['FUEL_TYPE'] == 'GAS')], values = ['COMPLETE_COUNT','ABORT_COUNT'],index = ['JOB_MONTH_DATE'],aggfunc=np.sum)
mnth_comp_total = pd.pivot_table(data = job_hist[job_hist['WORK_STATUS4'].isin(['CUSTOMER REFUSAL ABORT','TECHNICAL ABORT','NO ACCESS ABORT','INSTALL_AND_LEAVE','REVERT_TO_CLASSIC','COMPLETED'])], values = ['COMPLETE_COUNT','ABORT_COUNT'],index = ['JOB_MONTH_DATE'],aggfunc=np.sum)

```

```python
mnth_comp_df['SUCCESS_RATE'] = mnth_comp_df['COMPLETE_COUNT']/(mnth_comp_df['COMPLETE_COUNT']+mnth_comp_df['ABORT_COUNT'])
mnth_comp_elec['SUCCESS_RATE'] = mnth_comp_elec['COMPLETE_COUNT']/(mnth_comp_elec['COMPLETE_COUNT']+mnth_comp_elec['ABORT_COUNT'])
mnth_comp_gas['SUCCESS_RATE'] = mnth_comp_gas['COMPLETE_COUNT']/(mnth_comp_gas['COMPLETE_COUNT']+mnth_comp_gas['ABORT_COUNT'])
mnth_comp_total['SUCCESS_RATE'] = mnth_comp_total['COMPLETE_COUNT']/(mnth_comp_total['COMPLETE_COUNT']+mnth_comp_total['ABORT_COUNT'])

```

```python
mnth_df_time = mnth_df_time.rename(columns={"WORK_DURATION": "Dual Fuel Install Duration"})
mnth_elec_time = mnth_elec_time.rename(columns={"WORK_DURATION": "Electricity Install Duration"})
mnth_gas_time = mnth_gas_time.rename(columns={"WORK_DURATION": "Gas Install Duration"})
mnth_travel_time = mnth_travel_time.rename(columns={"ACTUAL_TRAVEL_TIME": "Average Travel Time"})

mnth_comp_df = mnth_comp_df.rename(columns={"SUCCESS_RATE": "Dual Fuel Success Rate"})
mnth_comp_elec = mnth_comp_elec.rename(columns={"SUCCESS_RATE": "Electricity Success Rate"})
mnth_comp_gas = mnth_comp_gas.rename(columns={"SUCCESS_RATE": "Gas Success Rate"})
mnth_comp_total = mnth_comp_total.rename(columns={"SUCCESS_RATE": "Total Success Rate"})

mnth_comp_df.drop(columns=['COMPLETE_COUNT','ABORT_COUNT'])
mnth_comp_elec.drop(columns=['COMPLETE_COUNT','ABORT_COUNT'])
mnth_comp_gas.drop(columns=['COMPLETE_COUNT','ABORT_COUNT'])
mnth_comp_total.drop(columns=['COMPLETE_COUNT','ABORT_COUNT'])

try:
    mnth_df_time.drop('2020-04-01',inplace=True)
except:
    pass
try:
    mnth_df_time.drop('2020-05-01',inplace=True)
except:
    pass
try:
    mnth_elec_time.drop('2020-04-01',inplace=True)
except:
    pass
try:
    mnth_elec_time.drop('2020-05-01',inplace=True)
except:
    pass
try:
    mnth_gas_time.drop('2020-04-01',inplace=True)
except:
    pass
try:
    mnth_gas_time.drop('2020-05-01',inplace=True)
except:
    pass
try:
    mnth_travel_time.drop('2020-04-01',inplace=True)
except:
    pass
try:
    mnth_travel_time.drop('2020-05-01',inplace=True)
except:
    pass


try:
    mnth_comp_df.drop('2020-04-01',inplace=True)
except:
    pass
try:
    mnth_comp_df.drop('2020-05-01',inplace=True)
except:
    pass
try:
    mnth_comp_elec.drop('2020-04-01',inplace=True)
except:
    pass
try:
    mnth_comp_elec.drop('2020-05-01',inplace=True)
except:
    pass
try:
    mnth_comp_gas.drop('2020-04-01',inplace=True)
except:
    pass
try:
    mnth_comp_gas.drop('2020-05-01',inplace=True)
except:
    pass
try:
    mnth_comp_total.drop('2020-04-01',inplace=True)
except:
    pass
try:
    mnth_comp_total.drop('2020-05-01',inplace=True)
except:
    pass


mnth_total = mnth_df_time.join(mnth_elec_time, how='left').join(mnth_gas_time, how='left').join(mnth_travel_time, how='left')
mnth_total_sr = mnth_comp_df.join(mnth_comp_elec, lsuffix='_left', rsuffix='_right', how='left').join(mnth_comp_gas, lsuffix='_left', rsuffix='_right', how='left').join(mnth_comp_total, lsuffix='_left', rsuffix='_right', how='left')
line_data_dur=pd.melt(mnth_total.reset_index(),id_vars='JOB_MONTH_DATE',value_vars=['Dual Fuel Install Duration','Electricity Install Duration','Gas Install Duration','Average Travel Time'])
line_data_sr=pd.melt(mnth_total_sr.reset_index(),id_vars='JOB_MONTH_DATE',value_vars=['Dual Fuel Success Rate','Electricity Success Rate','Gas Success Rate','Total Success Rate'])
line_data_dur
```

```python
mnth_mtr_type1 = job_hist[['JOB_MONTH_DATE','JOB_COUNT','INSTALLED_ELEC_SMT_TYPE']]
mnth_mtr_type2 = job_hist[job_hist['INSTALLED_GAS_SMT_FLAG'] =='Smart']



mnth_mtr_type2 = mnth_mtr_type2[['JOB_MONTH_DATE','JOB_COUNT','INSTALLED_GAS_MTR_TYPE']]
mnth_mtr_type1 = mnth_mtr_type1[(mnth_mtr_type1['INSTALLED_ELEC_SMT_TYPE'].notna() == True)]

mnth_mtr_type1['INSTALLED_ELEC_SMT_TYPE'] = mnth_mtr_type1['INSTALLED_ELEC_SMT_TYPE'].astype(str).str[:2]
mnth_mtr_type1['INSTALLED_ELEC_SMT_TYPE'].unique()
mnth_mtr_type2 = mnth_mtr_type2[(mnth_mtr_type2['INSTALLED_GAS_MTR_TYPE'].notna() == True)]
mnth_mtr_type1.rename(columns={'INSTALLED_ELEC_SMT_TYPE':'INSTALLED_SMT_TYPE'},inplace = True)
mnth_mtr_type2.rename(columns={'INSTALLED_GAS_MTR_TYPE':'INSTALLED_SMT_TYPE'},inplace = True)
mnth_mtr_type = pd.concat([mnth_mtr_type1,mnth_mtr_type2])
mnth_mtr_type = mnth_mtr_type.groupby(['JOB_MONTH_DATE','INSTALLED_SMT_TYPE']).sum().reset_index()
mnth_mtr_type.rename(columns={'JOB_COUNT':'INSTALLED_ASSETS'},inplace = True)
mnth_mtr_type = mnth_mtr_type.pivot(index='JOB_MONTH_DATE',columns='INSTALLED_SMT_TYPE',values='INSTALLED_ASSETS')
mnth_mtr_type.drop('2020-05-01',inplace=True)

```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

mnth_mtr_type[['S1','S2']].plot(kind='area',color = ["#5CC1CB","#b00402"],ax=ax1)

# Define Labels

ax1.set_xlabel('Year Month (YYYY-MM)', fontsize=16)
ax1.set_ylabel('Volume of Meter Installations, Higher is Better.', fontsize=16)

ax1.legend(title = 'Installed Meter Type')
# Twin Axes

plt.title("Dual Fuel Install Durations Overlayed on Installed Meter Type Volumes")
sns.lineplot(x='JOB_MONTH_DATE', y='value', hue='variable', palette=["#E3E000"], data=line_data_dur[line_data_dur['variable']=='Dual Fuel Install Duration'], linewidth = 3,ax=ax2)
ax2.set_ylabel('Install Time in Minutes, Lower is Better', fontsize=16)
ax2.legend(loc='upper left')
# Set limit 

ax2.set_ylim(100, 200)

# Display

plt.savefig("C:\Conda\Data Viz Assignment\Install_vol_dur_dualplot.png")
plt.show()
```

```python
corr_view = mnth_mtr_type[['S1','S2']].join(line_data_cor.set_index('JOB_MONTH_DATE'))
corr_view.rename(columns={"S1":"S1 Install Volumes","S2":"S2 Install Volumes"},inplace=True)
corr = corr_view.corr(method = 'spearman') 
sns.heatmap(corr,annot=True,cmap = 'coolwarm')
plt.title("Heatmap of Spearman Correlation Between Meter Type Install Volumes and Dual Fuel Install Duration",fontweight='bold', fontsize='16', horizontalalignment='center')
plt.tick_params(axis='x', direction='out', length=20, width=6)
plt.tick_params(axis='y', direction='out', length=20, width=6)
plt.savefig("C:\Conda\Data Viz Assignment\Correlation.png")
plt.show()
```
